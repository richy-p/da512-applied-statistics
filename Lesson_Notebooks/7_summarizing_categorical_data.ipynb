# Lesson 7 - Summarizing Categorical Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
cars = pd.read_csv('cars.csv')
```

```python
colors = cars['color']
colors.head()
```

```python
colors.nunique()
```

```python
colors.unique()
```

```python
colors.value_counts()
```

```python
colors.value_counts() / len(colors)
```

## Bar Chart

```python
bins = np.array(colors.value_counts().index)
bins
```

```python
counts = colors.value_counts().values
counts
```

```python
plt.rc('font',size=15)
fig,ax = plt.subplots(figsize=(12,7))
ax.bar(x=bins, height=counts)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(12,7))
ax.barh(y=bins, width=counts)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(12,7))
sns.barplot(y=bins, x=counts, palette='viridis')
plt.show()
```

```python
color_labels = bins.copy()
color_labels[7] = 'darkorchid'
```

```python
fig,ax = plt.subplots(figsize=(12,7))
ax.bar(x=bins, height=counts, color = color_labels, edgecolor='grey')
ax.grid(axis='x')
ax.set_title('Colors of Belarussian Used Cars for Sale')
ax.set_ylabel('Count')
ax.set_xlabel('Color')
fig.tight_layout()
```

```python
fig,ax = plt.subplots(figsize=(12,7))
sns.barplot(x=bins, y=counts, palette=color_labels, edgecolor='grey')
ax.set_title('Colors of Belarussian Used Cars for Sale')
ax.set_ylabel('Count')
ax.set_xlabel('Color')
fig.tight_layout()
```

## Pie Charts


```python
fig, ax = plt.subplots(figsize=(12,7))
plt.pie(x=counts, labels=bins, autopct='%.0f%%')
plt.show()
```

```python
def calc_number(perc):
    return round(perc / 100 * len(cars))
```

```python
fig, ax = plt.subplots(figsize=(12,7))
plt.pie(x=counts, labels=bins, autopct=calc_number, colors=color_labels)
plt.show()
```
